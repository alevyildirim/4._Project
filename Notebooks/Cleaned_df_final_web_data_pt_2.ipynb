```python
import pandas as pd

# Sample DataFrame with 'process_step' column
data = {'process_step': ['confirm', 'step_3', 'step_2', 'step_1', 'start']}
df = pd.DataFrame(data)

# Create dummy variables for 'process_step' column
dummies = pd.get_dummies(df['process_step'])

# Concatenate dummy variables with original DataFrame
df = pd.concat([df, dummies], axis=1)

# Drop the original 'process_step' column
df.drop('process_step', axis=1, inplace=True)

print(df)
```
